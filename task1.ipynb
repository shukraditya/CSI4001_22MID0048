### Word Cloud

```python
import nltk
from nltk.corpus import inaugural
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter
import string
```

```python
files = inaugural.fileids()
print(files)
```

```python
def create_inaugural_wordcloud():
    speech_corpus = ' '.join(inaugural.raw(fileid) for fileid in inaugural.fileids())
    
    tokens = word_tokenize(speech_corpus.lower())

    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens 
             if word not in stop_words 
             and word not in string.punctuation
             and word.isalnum()]

    word_freq = Counter(tokens)

    wordcloud = WordCloud(
        width=1600,
        height=800,
        background_color='white',
        max_words=100,
        contour_width=3,
        contour_color='steelblue'
    ).generate_from_frequencies(word_freq)
    
    # Display the word cloud
    plt.figure(figsize=(20,10))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title('US Presidential Inaugural Speech Word Cloud', pad=20)
    plt.show()
    
    return word_freq
```

```python
word_frequencies = create_inaugural_wordcloud()
```

### Frequency Analysis

```python
from nltk.probability import FreqDist, ConditionalFreqDist

def freq_analysis():
    word_frequencies = FreqDist()
    cfd = ConditionalFreqDist()
    stop_words = set(stopwords.words('english'))
    for fileid in inaugural.fileids():
        year=fileid.split('-')[0]
        speech = inaugural.raw(fileid)
        tokens = word_tokenize(speech.lower())
        clean_token = [
            word for word in tokens if word not in stop_words and word not in string.punctuation and word.isalnum()
        ]
        word_frequencies.update(clean_token)
        for word in clean_token:
            cfd[year][word] += 1
        return word_frequencies,cfd
```

```python
def display(word_frequencies,cfd):
    print("\n 25 most common words are:")
    for word,freq in word_frequencies.most_common(25):
        print(f"{word}: {freq}")

    print("\n Most common words by year are:")
    target_words = ['america','democracy','freedom','war','peace']
    years=sorted(cfd.conditions())

    print("\nYear",end="\t")
    for word in target_words:
        print(f"{word:>10}", end='\t')
    print()

    for year in years:
        print(year,end='\t')
        for word in target_words:
            print(f"{cfd[year][word]:>10}",end='\t')
        print()
    
```

```python
def analyze_bigrams():
    bigram_freq = FreqDist()
    
    for fileid in inaugural.fileids():
        speech = inaugural.raw(fileid)
        tokens = word_tokenize(speech.lower())
        
        # Create bigrams
        bigrams = list(nltk.bigrams(tokens))
        bigram_freq.update(bigrams)
    
    print("\nTop 20 most common bigrams:")
    for bigram, freq in bigram_freq.most_common(20):
        print(f"{bigram}: {freq}")
```

```python
word_freq, cfd = freq_analysis()
display(word_freq, cfd)
analyze_bigrams()
```

```python
def get_conditional_probability(cfd, condition, word):
    condition_total = sum(cfd[condition].values())
    word_count = cfd[condition][word]
    return word_count / condition_total if condition_total > 0 else 0
```
